# Ex 4

As I said multiple times in class, we want to learn how to solve problems with code, and not just how to use some pre-packed functions or modules. In this exercise you are required to solve a procedural problem that guitar builders have solved a long time ago: where to place the metal pieces on a neck, also called **frets**, in order to obtain the right note when the instrument is played. In the case of guitars, the note position is fixed (in violins, for example, the player decides where to press the string on the neck). The average lenght of the strings on a guitar is called "diapason". We can obtain the distance between the beginning of the diapason (fret nr. zero or nut) and first fret by dividing the diapson by the constant **17.817**. 

The guitar that you see in the picture has a diapson of 647 cm (or 25 and 1/2 inches); as a consenquence the distance between the nut and the first fret is `36.31363304708986` (you can try this yourself in a cell using the `/` operator). In order to obtain the distance between the 1st and the 2nd fret we have to subtract the previous number that we have just obtained from the diapason, to obtain a *new* diapason of `610.6863669529101`. If we divide this number by 17.817 again, we obtain the distance between the 1st fret and the 2nd fret. That will be the position of the first fret.
